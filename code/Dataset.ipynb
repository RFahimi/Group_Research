2-1 Reading the Files

```python
countries_raw = pd.read_excel("Countries.xlsx")
md_raw = pd.read_csv("MD per 10000.csv")
nurse_raw = pd.read_csv("Nursing and midwifery per 10000.csv")
life_raw = pd.read_csv("life_expectancy_all_countries.csv")
gdphexp_raw = pd.read_csv("GDP and health expenditure for  FCSL_data.csv")
gni_raw = pd.read_csv("GNI for FCSL_data.csv")
```

Step 3 Creating Countries List

```python
countries = [countries_raw.columns[0]] + countries_raw.iloc[:, 0].dropna().tolist()
countries
```

3-2 Uniformity of country names

```python
country_map = {
    "Congo, Dem. Rep.": "Congo, Democratic Republic of",
    "Democratic Republic of the Congo": "Congo, Democratic Republic of",
    "Congo, Rep.": "Congo, Republic of",
    "Republic of Congo": "Congo, Republic of",
    "Congo": "Congo, Republic of",
    "Venezuela": "Venezuela, RB",
    "Venezuela (Bolivarian Republic of)": "Venezuela, RB",
    "Yemen, Rep.": "Yemen, Republic of",
    "Yemen": "Yemen, Republic of",
    "West Bank and Gaza": "West Bank and Gaza (territory)",
    "Syrian Arab Rep.": "Syrian Arab Republic",
    "Timor Leste": "Timor-Leste",
    "Sao Tome and Principe": "São Tomé and Príncipe",
    "Micronesia, Fed. Sts.": "Micronesia, Federated States of",
    "Micronesia (Federated States of)": "Micronesia, Federated States of"
}

def clean_country(x):
    x = str(x).strip()
    return country_map.get(x, x)
```

Step 4 Preparing (Cleaning) "MD" file

```python
md = md_raw[["Location", "Period", "FactValueNumeric"]].copy()
md = md.rename(columns={
    "Location": "country",
    "Period": "year",
    "FactValueNumeric": "physicians_per_10000"
})

md["country"] = md["country"].apply(clean_country)
md["year"] = pd.to_numeric(md["year"], errors="coerce")
md["physicians_per_10000"] = pd.to_numeric(md["physicians_per_10000"], errors="coerce")

md = md[md["country"].isin(countries)].copy()

md["physicians_per_1000"] = md["physicians_per_10000"] / 10

md = md[["country", "year", "physicians_per_1000"]].copy()
md.head()
```

Step 5 Preparing (Cleaning) "Nursing" File

```python
nurse = nurse_raw[["Location", "Period", "FactValueNumeric"]].copy()

nurse = nurse.rename(columns={
    "Location": "country",
    "Period": "year",
    "FactValueNumeric": "nurses_midwives_per_10000"
})

nurse["country"] = nurse["country"].apply(clean_country)
nurse["year"] = pd.to_numeric(nurse["year"], errors="coerce")
nurse["nurses_midwives_per_10000"] = pd.to_numeric(nurse["nurses_midwives_per_10000"], errors="coerce")

nurse = nurse[nurse["country"].isin(countries)].copy()

nurse["nurses_midwives_per_1000"] = nurse["nurses_midwives_per_10000"] / 10

nurse = nurse[["country", "year", "nurses_midwives_per_1000"]].copy()

nurse.head()
```

Step 6 Preparing (Cleaning) "life expectancy" File

```python
life_raw.head()
```

```python
life_raw.columns
```

```python
Index(['economy', 'Country', 'YR2000', 'YR2001', 'YR2002', 'YR2003', 'YR2004',
       'YR2005', 'YR2006', 'YR2007', 'YR2008', 'YR2009', 'YR2010', 'YR2011',
       'YR2012', 'YR2013', 'YR2014', 'YR2015', 'YR2016', 'YR2017', 'YR2018',
       'YR2019', 'YR2020'],
      dtype='object')
```

```python
life = life_raw.copy()

life["Country"] = life["Country"].apply(clean_country)

life = life[life["Country"].isin(countries)].copy()

year_cols = [c for c in life.columns if str(c).startswith("YR")]

life = life.melt(
    id_vars=["Country"],
    value_vars=year_cols,
    var_name="year",
    value_name="life_expectancy"
)

life = life.rename(columns={"Country": "country"})

life["year"] = life["year"].str.replace("YR", "", regex=False)
life["year"] = pd.to_numeric(life["year"], errors="coerce")

life["life_expectancy"] = pd.to_numeric(life["life_expectancy"], errors="coerce")

life = life[["country", "year", "life_expectancy"]].copy()

life.head()
```

```python
len(life["country"].unique())
```

```python
sorted(life["country"].unique())
```

Step 7 Preparing (Cleaning) "GNI" File

```python
gni_raw.head()
```

```python
gni_raw.columns
```

```python
gni = gni_raw.copy()

gni["Country Name"] = gni["Country Name"].apply(clean_country)

gni = gni[gni["Country Name"].isin(countries)].copy()

year_cols = [c for c in gni.columns if "[YR" in str(c)]

gni = gni.melt(
    id_vars=["Country Name"],
    value_vars=year_cols,
    var_name="year",
    value_name="gni_ppp_current_international"
)

gni = gni.rename(columns={"Country Name": "country"})

gni["year"] = gni["year"].str.extract(r"(\d{4})")
gni["year"] = pd.to_numeric(gni["year"], errors="coerce")

gni["gni_ppp_current_international"] = pd.to_numeric(
    gni["gni_ppp_current_international"].replace("..", pd.NA),
    errors="coerce"
)

gni = gni[["country", "year", "gni_ppp_current_international"]].copy()

gni.head()
```

```python
len(gni["country"].unique())
```

```python
sorted(gni["country"].unique())
```

Step 8 Preparing (Cleaning) "Health Expenditure" and "GDP" Files

```python
gdphexp_raw["Series Name"].drop_duplicates().tolist()
```

```python
gdp_part = gdphexp_raw[gdphexp_raw["Series Name"] == "GDP per capita (current US$)"].copy()

hexp_part = gdphexp_raw[gdphexp_raw["Series Name"] == "Current health expenditure per capita (current US$)"].copy()
```

Step 9 Reshaping GDP from wide format to long format

```python
gdp_part["Country Name"] = gdp_part["Country Name"].apply(clean_country)
gdp_part = gdp_part[gdp_part["Country Name"].isin(countries)].copy()

year_cols = [c for c in gdp_part.columns if "[YR" in str(c)]

gdp = gdp_part.melt(
    id_vars=["Country Name"],
    value_vars=year_cols,
    var_name="year",
    value_name="gdp_per_capita_current_usd"
)

gdp = gdp.rename(columns={"Country Name": "country"})

gdp["year"] = gdp["year"].str.extract(r"(\d{4})")
gdp["year"] = pd.to_numeric(gdp["year"], errors="coerce")

gdp["gdp_per_capita_current_usd"] = pd.to_numeric(
    gdp["gdp_per_capita_current_usd"].replace("..", pd.NA),
    errors="coerce"
)

gdp = gdp[["country", "year", "gdp_per_capita_current_usd"]].copy()

gdp.head()
```

```python
len(gdp["country"].unique())
```

```python
sorted(gdp["country"].unique())
```

Step 10 Reshaping Health expenditure from wide format to long format

```python
hexp_part["Country Name"] = hexp_part["Country Name"].apply(clean_country)
hexp_part = hexp_part[hexp_part["Country Name"].isin(countries)].copy()

year_cols = [c for c in hexp_part.columns if "[YR" in str(c)]

hexp = hexp_part.melt(
    id_vars=["Country Name"],
    value_vars=year_cols,
    var_name="year",
    value_name="health_expenditure_per_capita_current_usd"
)

hexp = hexp.rename(columns={"Country Name": "country"})

hexp["year"] = hexp["year"].str.extract(r"(\d{4})")
hexp["year"] = pd.to_numeric(hexp["year"], errors="coerce")

hexp["health_expenditure_per_capita_current_usd"] = pd.to_numeric(
    hexp["health_expenditure_per_capita_current_usd"].replace("..", pd.NA),
    errors="coerce"
)

hexp = hexp[["country", "year", "health_expenditure_per_capita_current_usd"]].copy()

hexp.head()
```

```python
len(hexp["country"].unique())
```

Step 11 Creating Main Chart

```python
years = list(range(2000, 2025))

panel = pd.DataFrame(
    list(itertools.product(countries, years)),
    columns=["country", "year"]
)
```

Step 12 Merging All

```python
final = panel.merge(md, on=["country", "year"], how="left")
final = final.merge(nurse, on=["country", "year"], how="left")
final = final.merge(life, on=["country", "year"], how="left")
final = final.merge(gdp, on=["country", "year"], how="left")
final = final.merge(hexp, on=["country", "year"], how="left")
final = final.merge(gni, on=["country", "year"], how="left")

final.head()
```
